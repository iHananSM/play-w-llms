```python
from transformers import pipeline
import datetime

def get_schedule(projects, progress, deadlines, work_hours, energy_level):
    """
    Generates a daily schedule using a Hugging Face model.
    """
    prompt = f"""
    You are a productivity assistant for a researcher. Today is {datetime.date.today()}.
    The user has {work_hours} hours available for work.
    Their energy level is {energy_level}.
    They are working on these projects:
    {projects}
    
    Current progress:
    {progress}
    
    Deadlines:
    {deadlines}
    
    Suggest a structured daily schedule that prioritizes deep focus work in the morning and lighter tasks later.
    """
    
    generator = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct")  # Adjust model as needed
    response = generator(prompt, max_length=500, do_sample=True, temperature=0.7)
    
    return response[0]['generated_text']

# Example input
projects = {
    "Health4AI": "Paper writing stage (50% done)",
    "LLM Evaluation": "Paper writing stage (30% done)",
    "AI+Chem": "Evaluation stage",
    "Chatbot": "Initial development",
    "Arabic Sci LLM": "Data collection stage"
}

deadlines = {
    "Health4AI": "April 15, 2025",
    "LLM Evaluation": "April 20, 2025",
    "AI+Chem": "May 10, 2025",
    "Chatbot": "June 1, 2025",
    "Arabic Sci LLM": "Ongoing"
}

work_hours = 5  # Adjust based on Ramadan/normal schedule
energy_level = "Medium"  # Options: High, Medium, Low

# Generate schedule
schedule = get_schedule(projects, projects, deadlines, work_hours, energy_level)
print("\nGenerated Schedule:\n", schedule)
```
